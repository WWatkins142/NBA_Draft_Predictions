```python
import pandas as pd
import csv, json
```

```python
# read in csv file
draft = pd.read_csv("DraftProbabilities.csv")
clean2022 = pd.read_csv("Cleaned2022.csv")
```

```python
draft.head()
```

```python
clean2022.head()
```

```python
# add columns to draft dataframe
# "GP", "Min_per", "Ortg", "TS_per", "AST_per", "TO_per", "blk_per", "FT_per", "porpag", "adjoe"
draft["Games Played"] = clean2022["GP"]
draft["Minutes Percentage"] = clean2022["Min_per"]
draft["Offensive Rating"] = clean2022["Ortg"]
draft["True Shooting Percentage"] = clean2022["TS_per"]
draft["Assist Percentage"] = clean2022["AST_per"]
draft["Turnover Percentage"] = clean2022["TO_per"]
draft["Block Percentage"] = clean2022["blk_per"]
draft["Free Throw Percentage"] = clean2022["FT_per"]
draft["Points Over Replacement Per Adjusted Game"] = clean2022["porpag"]
draft["Adjusted Offensive Efficiency"] = clean2022["adjoe"]
draft.head()
```

```python
# creating csv from possible drafted 2022 players dataframe
draft.to_csv("Complete2022Draft.csv")
```

```python
# file paths
csvDraft = 'Complete2022Draft.csv'
jsonDraft = 'draft.json'
```

```python
# read csv file and add to data
data = {}
with open(csvDraft) as  csvFile:
    csvReader = csv.DictReader(csvFile)
    for rows in csvReader:
        id = rows['id']
        data[id] = rows
```

```python
# create json file and write data on it
with open(jsonDraft, 'w') as jsonFile:
    # pretty data
    jsonFile.write(json.dumps(data, indent=4))
```
