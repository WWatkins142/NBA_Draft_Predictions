```python
# depencies
import pandas as pd
import os 
import csv
import numpy as np
```

```python
#Import files
collegeBball = "Resources/CollegeBasketballPlayers2009-2021.csv"
currcollegeBball = "Resources/CollegeBasketballPlayers2022.csv"
draftPlayers = "Resources/DraftedPlayers2009-2021.csv"
```

```python
#College Players 2009-2021 Dataframe
pastColPlayersDF = pd.read_csv(collegeBball)
pastColPlayersDF.head(3)
```

```python
# read files and create College basketball players 2022 df
currColPlayersDF = pd.read_csv(currcollegeBball)
currColPlayersDF.head()
```

```python
# read files and create drafted players 2009-2021 df
draftPlayersDF = pd.read_csv(draftPlayers)
draftPlayersDF.head()
```

```python
# rename player column in draftPlayersDF
draftPlayersDF = draftPlayersDF.rename(columns={"PLAYER":"player_name","TEAM":"nba team","AFFILIATION":"team","YEAR":"year", "OVERALL PICK":"pick"})
draftPlayersDF.head(3)
```

```python
pastColPlayersDF['pick'].fillna("Undrafted",inplace= True)
pastColPlayersDF.head()
```

```python
draftPlayersDF['pick'].fillna("Undrafted",inplace= True)
draftPlayersDF['pick'].head(50)
```

```python
# replacing numeric values for boolean values
pastColPlayersDF.loc[pastColPlayersDF['pick'] !="Undrafted", 'pick'] = "Drafted"
pastColPlayersDF['pick'].head(50)
```

```python
# replacing numeric values for boolean values
draftPlayersDF.loc[draftPlayersDF['pick'] !="Undrafted", 'pick'] = "Drafted"
draftPlayersDF['pick'].head(50)
```

```python
# Join draft players df 09-21 and college basketball players 09-21 df on player name,team, and year
newdraftDF = pd.merge(pastColPlayersDF,draftPlayersDF,on=["player_name","pick"], how='outer')
newdraftDF
```

```python
# check for and drop null values in college Players 2022
#currColPlayersDF['team'].isna().sum()
```

```python
#drop pid column from newdraftDF 
newdraftDF = newdraftDF.drop(columns='pid')
newdraftDF.head(3)
```

```python
# drop pid column from current college players df
potentialdraftDF = currColPlayersDF.drop(columns='pid')
potentialdraftDF.head(3)
```

```python
#update the pick column to draft status
newdraftDF = newdraftDF.rename(columns={"pick":"draft_status"})
newdraftDF.head(2)
```

```python
#update the pick column to draft status

potentialdraftDF = potentialdraftDF.rename(columns={"pick":"draft_status"})
potentialdraftDF.head(3)
```

```python
#  replace all null values in the pick column of the potentialdrafDF with "not drafted"
potentialdraftDF['draft_status'].fillna("Undrafted",inplace= True)
potentialdraftDF['draft_status'].isna().sum()
```

```python
# export cleaned newdraft dataframe
newdraftDF.to_csv('Resources/Updated_new_draft.csv')
```

```python
# export cleaned potentialdraft dataframe 
potentialdraftDF.to_csv('Reources/Potential_Draft.csv')
```
